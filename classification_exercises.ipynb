```python
# grab the iris data

def get_iris_data():
    sql_query = """
    SELECT species_id,
    species_name,
    sepal_length, 
    sepal_width, 
    petal_length,
    petal_width
    FROM measurements
    JOIN species
    USING (species_id)
    """
    return pd.read_sql(sql_query, get_connection("iris_db"))
```
